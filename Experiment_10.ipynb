```python
# Create DataFrame
data={
    "Name":["A","B","C"],
    "Marks":[85,90,78]
}
df = pd.DataFrame(data)
print(df)
# df.head() gives us the first 5 rows
# df.tail() gives us the last 5 row
```

```python
# Find the Structure of the DataFrame
print(df.shape)
```

```python
print(df.ndim)
```

```python
# Display Column Names
print(df.columns)
```

```python
# Check Data Types of each Column
print(df.dtypes)
```

```python
# Access Individual Column
print(df["Name"])
```

```python
print(df["Marks"])
```

```python
# Access Name of first row
print(df.loc[0,"Name"])
```

```python
print(df.iloc[0,1])
```

```python
# Add a New Column
'''If marks 80-90 First Class
if marks>=90-Distinction
if marks>=80-Second Class'''

df["Grade"]=["First Class","Distinction","Second Class"]
print(df)
```

```python
# Update Data in DataFrame
# Change marks of student A to 88.
df.loc[0,"Marks"]=88
df.iloc[1,1]=91
print(df)
```

```python
df1=df.drop("Grade", axis=1)
print(df1)
print(df)
```

```python
# Perform Basic Statistical Analysis
mean1=df["Marks"].mean()
print(mean1)
```

```python
min1=df["Marks"].min()
print(min1)
```

```python
max1=df["Marks"].max()
print(max1)
```

```python
# Apply Condition (Filtering)
# Display students scoring more than 80.
df[df["Marks"]>80]
```
